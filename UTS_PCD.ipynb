```python
import cv2 as cv
from google.colab import files
from google.colab.patches import cv2_imshow

# Deteksi wajah menggunakan Haar Cascade
def detect_smallest_face():
    uploaded = files.upload()

    for image_path in uploaded.keys():
        img = cv.imread(image_path)
        gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

        # Inisialisasi kelas Haar Cascade untuk deteksi wajah
        face_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')

        # Deteksi wajah
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=6)


        # Temukan wajah terkecil
        smallest_face = min(faces, key=lambda face: face[2]*face[3])

        # Gambar wajah terkecil dengan warna merah
        if len(smallest_face) > 0:
            x, y, width, height = smallest_face
            cv.rectangle(img, (x, y), (x + width, y + height), (0, 0, 255), thickness=2)

        # Tampilkan gambar yang telah ditandai
        cv2_imshow(img)

# Panggil fungsi untuk mendeteksi wajah terkecil
detect_smallest_face()
```

```python
import cv2 as cv
from google.colab import files
from google.colab.patches import cv2_imshow

# Deteksi wajah menggunakan Haar Cascade
def detect_largest_face():
    uploaded = files.upload()

    for image_path in uploaded.keys():
        img = cv.imread(image_path)
        gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

        # Inisialisasi kelas Haar Cascade untuk deteksi wajah
        face_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')

        # Deteksi wajah
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=6)

        # Inisialisasi variabel untuk wajah terbesar
        largest_face = None
        largest_area = 0

        for (x, y, width, height) in faces:
            # Gambar kotak di sekitar wajah
            cv.rectangle(img, (x, y), (x + width, y + height), (0, 255, 0), thickness=2)

            # Hitung luas wajah saat ini
            current_area = width * height

            # Periksa apakah luas saat ini lebih besar dari luas wajah terbesar yang ada
            if current_area > largest_area:
                largest_face = (x, y, width, height)
                largest_area = current_area

        # Gambar wajah terbesar dengan warna merah
        if largest_face is not None:
            x, y, width, height = largest_face
            cv.rectangle(img, (x, y), (x + width, y + height), (0, 0, 255), thickness=2)

        # Tampilkan gambar yang telah ditandai
        cv2_imshow(img)

# Panggil fungsi untuk mendeteksi wajah terbesar
detect_largest_face()
```
