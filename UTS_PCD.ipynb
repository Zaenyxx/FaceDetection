```python
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
!ls FaceDetection/img
```

```python
img = cv.imread('FaceDetection/img/foto1.jpg')
cv2_imshow(img)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
haar_cascade = cv.CascadeClassifier('uts-pcd-face-detection/haar_face.xml')

faces = haar_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=6)

print(f'Jumlah wajah yang terdeteksi = {len(faces)}')

for (x,y,width,height) in faces:
    cv.rectangle(img, (x,y), (x+width,y+height), (0,255,0), thickness=2)
```
